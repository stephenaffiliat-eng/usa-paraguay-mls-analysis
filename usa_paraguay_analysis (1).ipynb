# USA vs Paraguay — MLS Player Impact Analysis
This notebook reproduces the analysis and visuals used in the Substack article.


```python
import pandas as pd

data = {
    "player": ["Matthew Freese", "Tim Ream", "Sebastian Berhalter"],
    "minutes": [90, 90, 45],
    "accurate_passes": [12, 84, 20],
    "total_passes": [20, 90, 28],
    "passes_opp_half_acc": [0, 38, 10],
    "passes_opp_half_total": [8, 43, 17],
    "passes_own_half_acc": [12, 46, 10],
    "passes_own_half_total": [12, 47, 11],
    "long_balls_acc": [2, 7, None],
    "long_balls_total": [10, 9, None],
    "recoveries": [6, 5, 3],
    "def_contributions": [0, 6, 1],
    "ground_duels_won": [0, 2, 1],
    "ground_duels_total": [0, 2, 1],
    "aerial_duels_won": [None, 3, 2],
    "aerial_duels_total": [None, 1, 2],
    "possession_lost": [8, 7, 10],
    "xG": [0, 0, 0.02],
    "xA": [0, 0.03, 0.12],
    "shots": [0, 0, 1],
}

df = pd.DataFrame(data)

df["pass_accuracy"] = df["accurate_passes"] / df["total_passes"] * 100
df["opp_half_accuracy"] = df["passes_opp_half_acc"] / df["passes_opp_half_total"] * 100
df["own_half_accuracy"] = df["passes_own_half_acc"] / df["passes_own_half_total"] * 100
df["long_ball_accuracy"] = df["long_balls_acc"] / df["long_balls_total"] * 100
df["possessions_per_min"] = df["possession_lost"] / df["minutes"]

df
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))
sns.heatmap(
    df.set_index("player")[[
        "accurate_passes","total_passes","recoveries","def_contributions",
        "ground_duels_won","aerial_duels_won","possession_lost","shots"
    ]],
    annot=True,
    cmap="coolwarm"
)
plt.title("Full Metric Heatmap — USA vs Paraguay")
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.bar(df["player"], df["pass_accuracy"], color=["steelblue","darkgreen","orange"])
plt.ylabel("Pass accuracy (%)")
plt.title("Pass accuracy — USA vs Paraguay")
plt.ylim(0, 100)
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.bar(df["player"], df["def_contributions"], color=["steelblue","darkgreen","orange"])
plt.ylabel("Defensive contributions (tackles + interceptions + blocks + clearances)")
plt.title("Defensive impact — USA vs Paraguay")
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# number of touches
n = 102

# generate synthetic coordinates based on Ream's real heat map profile
x = np.random.normal(loc=30, scale=8, size=n)
y = np.random.normal(loc=45, scale=12, size=n)

# clip to pitch boundaries
x = np.clip(x, 0, 105)
y = np.clip(y, 0, 68)

df_heat = pd.DataFrame({"x": x, "y": y})

plt.figure(figsize=(10,6))
sns.kdeplot(
    data=df_heat,
    x="x",
    y="y",
    fill=True,
    cmap="Reds",
    thresh=0.05,
    bw_adjust=0.8
)

plt.title("Tim Ream — Synthetic Heat Map (USA vs Paraguay)")
plt.gca().invert_yaxis()
plt.xlabel("Pitch Length")
plt.ylabel("Pitch Width")
plt.show()
```
